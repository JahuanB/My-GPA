```python
def grade_to_gpa(grade):
    # Define the GPA mapping
    gpa_scale = {
        'A': 4.0,
        'A-': 3.7,
        'B+': 3.3,
        'B': 3.0,
        'B-': 2.7,
        'C+': 2.3,
        'C': 2.0,
        'C-': 1.7,
        'D+': 1.3,
        'D': 1.0,
        'F': 0.0
    }

    # Return the corresponding GPA value
    return gpa_scale.get(grade, 0.0)

def calculate_gpa(grades):
    total_points = 0
    total_courses = len(grades)

    for grade in grades:
        total_points += grade_to_gpa(grade)

    gpa = total_points / total_courses if total_courses > 0 else 0
    return round(gpa, 1)


grades = ['A-', 'B+', 'B+', 'B+', 'B','A-','A','B','B','C','A-','B']
gpa = calculate_gpa(grades)
print(f"Your GPA is: {gpa}")
```
